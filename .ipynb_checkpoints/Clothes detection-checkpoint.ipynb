# Нейронная сеть для обнаружения и классификации категории одежды по фотографии

````python
#Импорт библиотеки fastai

from fastai import *
from fastai.vision import *
````

### Создание датасета из Google Images

Изображения скачиваются из Google Images с помощью скрипта:
```javascript
urls = Array.from(document.querySelectorAll('.rg_di .rg_meta')).map(el=>JSON.parse(el.textContent).ou);
window.open('data:text/csv;charset=utf-8,' + escape(urls.join('\n')));
```
Далее создаются директории с соответсвующими файлами с ссылками на изображения

````python
folder = 'tshirts'
file = 't-shirts.txt'
````

````python
folder = 'shoes'
file = 'shoes.txt'
````

````python
folder = 'dress'
file = 'dress.txt'
````

````python
folder = 'coat'
file = 'coat.txt'
````

````python
folder = 'trousers'
file = 'trousers.txt'
````

````python
path = Path('data')
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
````

````python
path.ls()
````

````python
classes = ['tshirts','shoes','dress','coat','trousers']
````

Модель будет классфицироваться на 5 классов: 
* Футболка
* Обувь
* Платье
* Пальто
* Брюки

### Скачивание изображений на сервер

````python
download_images(path/folder/file, dest, max_pics=200)
````

````python
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
````

````python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
````

````python
data.classes
````

Далее произвольный набор изображений с помеченными тегами

````python
data.show_batch(rows=3, figsize=(7,8))
````

````python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
````

### Создание и обучение модели

````python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
````

````python
learn.fit_one_cycle(4)
````

Ошибка после 3 эпохи составляет 0,06.

````python
learn.save('stage-1')
````

### Создание матрицы ошибок

````python
interp = ClassificationInterpretation.from_learner(learn)
````

````python
interp.plot_confusion_matrix()
````

Не удивительно, что чаще всего ошибки в классификации происходили между изображениями с пальто и платьями. В остальных случаях за некоторым исключением модель работает эффективно.
